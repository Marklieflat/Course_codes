```python
import numpy as np
import pandas as pd
import random
import datetime as dt
```

```python
data = np.loadtxt('seeds_dataset.txt')
data_x = data[:,0:7]
k = 3
```

```python
def distance(node1,node2):
    return np.sum((node1-node2)**2)**0.5
```

```python
def init_node(data,k):
    data = list(data)
    init_node_ = random.sample(data,k)
    return init_node_
```

```python
def assign_clusters(data,current_centroids):
    current_clusters = dict()
    for i in data:
        current_min_dist = 1000000
        cluster_index = -1
        for j in range(k):
            centroid = current_centroids[j]
            current_dist = distance(i,centroid)
            if current_dist < current_min_dist:
                current_min_dist = current_dist
                cluster_index = j
        if cluster_index not in current_clusters.keys():
            current_clusters[cluster_index] = []
        current_clusters[cluster_index].append(i)
    return current_clusters
```

```python
def assign_centroids(current_clusters):
    new_centroids = []
    for cluster_index in range(k):
        new_centroid = np.mean(current_clusters[cluster_index],axis = 0)
        new_centroids.append(new_centroid)
    return new_centroids
```

```python
start = dt.datetime.now()
centroids = init_node(data,k)
current_clusters = assign_clusters(data,centroids)
iteration = 0
while not np.array_equal(centroids,assign_centroids(current_clusters)):
    centroids = assign_centroids(current_clusters)
    current_clusters = assign_clusters(data,centroids)
    iteration += 1
print(centroids)
print("cluster_0 have",current_clusters[0])
print("cluster_1 have",current_clusters[1])
print("cluster_2 have",current_clusters[2])

end = dt.datetime.now()
elapse = end - start
print(elapse)
print(iteration)
```

```python
def Silhouette_coefficient(current_clusters,centroids):
    s_coef = 0
    for i in range(k):
        current_min_dist = 1000000
        for i_ in range(k):
            if i != i_:
                current_dist = distance(i,i_)
                if current_dist < current_min_dist:
                    current_min_dist = current_dist
                    nearest_i = i_
        for j in current_clusters[i]:
            dist_j = 0
            for z in current_clusters[i]:
                dist_j += distance(j,z)
            a = dist_j/len(current_clusters[i])
            for z_ in current_clusters[nearest_i]:
                dist_j += distance(j,z_)
            b = dist_j/len(current_clusters[i])
            s = (b-a) / max([a,b])
            s_coef += s
    s_coef = s_coef/210
    return s_coef
Silhouette_coefficient(current_clusters,centroids)
```
