```python
import pandas as pd
import matplotlib.pyplot as plt 
from matplotlib.pyplot import MultipleLocator
```

```python
df = pd.read_csv(r'D:\Code Library\FIN3380_summer\Project_1\PORT_CRET.csv')
df
```

```python
plt.figure(figsize = [20,8])
plt.plot(df.date, df.P1)
plt.plot(df.date, df.P2)
plt.plot(df.date, df.P3)
plt.plot(df.date, df.P4)
plt.plot(df.date, df.P5)
plt.plot(df.date, df.P6)
plt.plot(df.date, df.P7)
plt.plot(df.date, df.P8)
plt.plot(df.date, df.P9)
plt.plot(df.date, df.P10)
plt.xlabel('Backtesting time series(1980-2021)')
plt.ylabel('Cumulative return')
plt.xticks(rotation = 45)
plt.title('Portfolios Cumulative Returns over 1980-2021')
x_major_locator=MultipleLocator(12)
ax=plt.gca()
ax.xaxis.set_major_locator(x_major_locator)
plt.grid(axis = 'y')
plt.legend()
plt.show()
```
