# Event Study: A Demonstration
- **Efficient Market Hypothesis** summarized in Fama (1970) provides a theoretical ground for empirical researches in Financial Economics, Accounting, etc. to investigate the informational component in asset pricing.
- **Event studies** are often used in the applications that are related to semi-strong form efficiency (information become publicly available) or strong form efficiency (private information), such as

  - Whether a given type of publicly available information might an investor use to generate higher than normal returns?

  - How much time is required for a given type of information to be reflected in asset prices?

In a typical event study, stock return behavior is analyzed in a relative short period of time, such as a few days surrounding the announcement date. An event could be firms' specific news (e.g., M&A and Earning Announcement) or macroeconomic news (e.g., FOMC announcement). The goal is to measure the information impact, abnormal return, due to the event which is the difference between the stock’s actual return and a proxy for the stock’s return in the absence of the event. 
- a pricing model such as the Market(CAPM) or Fama-French three factor model 
- a broad market index 
- a firm characteristic-based portfolio

I am developing a step-by-step Python case study to demonstrate how to perform a typical event study to identify previously unforeseen phenomena first described by Shleifer (1986) and Harris and Gurel (1986). Please see the [this link](https://wrds-www.wharton.upenn.edu/pages/support/applications/python-replications/programming-python-wrds-event-study/) for a full-fledged Python API at WRDS. 

## Background
Asset pricing models based efficient markets hypothesis (e.g., Fama and French 1993) tend to assume nearly horizontal demand curves for stocks, i.e. investors are price takers. Shleifer (1986) and Harris and Gurel (1986) document a remarkable fact: when a stock was added to the index, it jumps in price and much of this jump is permanent. The fact that a stock jumps in value upon inclusion may indicate that the stock price can be changed through an increased demand (e.g., from index funds) even though its fundamental value does not change. 

Constituents of the S&P 500 index can be taken out of the index because of a merger or bankruptcy and are replaced by other firms. Standard and Poor emphasizes that in selecting stocks for inclusion, they are simply trying to make their index representative of the US economy, not to convey any information about the level or riskiness of a firm's future cash flows 
> S&P states six criteria for inclusion: size, industry classification, capitalization, trading volume, emerging companies/industries, and responsiveness of the movements of stock price to changes in industry affairs, Those criteria are all public information and none of them is concerned with the future performance of the firm.

## Trading Days Calendar

To statistically value corporate events, it is essential to align stock price fluctuations around the event days, when these events are made public and commonly referred to as "day 0" in event studies.
![TimeLine](http://drive.google.com/uc?export=download&id=1n-_eX9-IQYVz7rBbfzOtcrvV_HlUttWd) 
Also, there are more than two non-trading days (including holidays) each week in the United States, so **using calendar days** to align daily stock price movements with event days is **NOT** optimum. Therefore, event studies require trading day calendars corresponding to the events and the relative trading days to the events.

## Event Study FrameWork

An event study is a method of isolating the portion of a stock price movement due to a specific event by benchmarking a proxy for the stock's return in the absence of the event. 
![EventStudy](http://drive.google.com/uc?export=download&id=1GAVrbs6DzNvIwuyYA4FspRea3zFBgDeK) 

The isolated price movement is referred as the tock i’s Abnormal Return, $AR_{i,t}$ is defined as the deviation of its return from its normal return $NR_{i,t}$.

>
> $\widehat{{AR_{i,t}}}=R_{i,t}-\widehat{NR_{i,t}}$
>
- **Market Adjusted Model**: Use market index as benchmark 
- **Market Model (Not Exact CAPM)**: $\widehat{{AR_{i,t}}}=R_{i,t}-(\hat{a_i}+\hat{b_i}R_{m,t})$
- **Fama & French Model**: $\widehat{{AR_{i,t}}}=R_{i,t}-（R_{f,t}+\hat{a_i}+\hat{b_i}(R_{m,t}-R_{f,t})+\hat{s_i}SMB+\hat{h_i}HML)$
- **Fama & French + Momentum**: $\widehat{{AR_{i,t}}}=R_{i,t}-（R_{f,t}+\hat{a_i}+\hat{b_i}(R_{m,t}-R_{f,t})+\hat{s_i}SMB+\hat{h_i}HML++\hat{u_i}UMD)$

To separate abnormal price movement caused by event information, abnormal returns generated by regression models must estimate coefficients in "regular" days.

![TimeLine](http://drive.google.com/uc?export=download&id=19d-_l6I4XpNx4hojknc-j0pC4dOURTCV)

Cumulative Abnormal Returns (CAR) from relative day s to t is estimated as 
>
> $CAR(s,t)=\sum_{s}^t{AR_{i,t}}$
>

> DSI table in CRSP Database contains index returns in trading days. The following block is to obtain trading day calendar and keep relative days (named as CRSPDAY) to the first trading day in CRSP, Dec 31, 1925 ![TradingCalendar](http://drive.google.com/uc?export=download&id=1_LHRC0rsvKjScZ7kpoScHtSXobofNoFx) 

```python
crspday = wrds_sql.raw_sql("""
    select distinct date
    from crsp.dsi
    order by date;
        """)
crspday=crspday.reset_index()

crspday.columns=['crspday', 'date']
crspday.head()
```

> DSP500LIST Table in CRSP Database contains PERMNO (permanent Security ID), inclusion date, and exlusion date (if available) for S&P500 constituents. We focus on replication of Shleifer (1986), in which the sample period is from September 1, 1976 to December 31, 1983. 

```python
#S&P500 Inclusion Events
event = wrds_sql.raw_sql(""" 
    select distinct permno, start as edate 
    from crsp.dsp500list 
    where start >= '{}' and start <= '{}'
    """.format(datetime(1990,1,1).date(), datetime(2020,12,31).date()))
```

> The following SQL statement advances the event date to the first trading day on and succeeding the event date. For example, if a stock is hypothetically added to the S&P 500 on July 4, 2022, we would consider July 5, 2022, the event day since that is the first post-event day when investors might ordinarily trade on the event-related information.

```python
crspday.to_sql('crspday', local_sql, index=False, if_exists='replace') 
event.to_sql('event', local_sql, index=False, if_exists='replace')  

sqlcode="""
    select distinct a.*, crspday 
    from event as a left join crspday as b
    on edate<=date and date<=date(edate, "4 day") 
    group by a.permno, edate
    having date=min(date) 
"""
event=pd.read_sql_query(sqlcode, local_sql)
```

```python
estper=200;                       #Length of the estimation window in trading days                       
pos=60;                           #End of the event window (greater than CAR_to)                         
pre=-30;                          #Beginning of Abnormal Return Window(less than CAR_from for a est.GAP) 
minest=120;                       #Minimum of non-missing returns required for estimation                
CAR_from=0;                       #Beginning of event window of interest                                 
CAR_to=0;                         #End of event window of interest                                       
evtwin=CAR_to-CAR_from + 1;       #length of event window in trading days 
```

```python
event.to_sql('event', local_sql, index=False, if_exists='replace')  
```

```python
sqlcode="""
    select distinct a.*, 
    d.date as estper_beg, /*Start of the Estimation Window*/
    e.date as estper_end, /*End of the Estimation Window*/
    b.date as evtwin_beg, /*Start of the Abnormal Return Window*/
    a.edate as evtdate, /*Event Date*/
    c.date as evtwin_end /*End of the Abnormal Return Window*/
    from event as a inner join crspday as b
    on b.crspday-a.crspday= {} /*pre*/
    inner join crspday as c
    on c.crspday-a.crspday= {} /*pos*/
    inner join crspday as d
    on b.crspday-d.crspday= {} /*estper*/
    inner join crspday as e
    on b.crspday-e.crspday=1
""".format(pre, pos, estper)
tradedates=pd.read_sql_query(sqlcode, local_sql)
```

> For each S&P500 inclusion event, we select calendard dates for estimation windows and event windows with specific numbers of trading days relative to event day. 

```python
tradedates
```

> We select all the PERMNOs that are included into S&P500 in our sample period and obtain daily returns from CRSP for corresponding stocks. In addition, we add market return, risk-free return, SMB, HML, and UMD for the later regressions. 

```python
permno_tuple = tuple(tradedates['permno'])

daily_rets = wrds_sql.raw_sql("""
    select distinct permno, a.date, a.ret, ret-rf as retrf, mktrf, smb, hml, umd, 
    mktrf+rf as mkt, ret- mktrf - rf as exret /*Market-adjusted Return*/
    from crsp.dsf as a inner join ff.Factors_daily as b
    on a.date=b.date and permno in {} /*permno list*/
    order by a.permno, a.date;
    """.format(permno_tuple))
```

```python
daily_rets.tail(3)
```

> Then, we match daily returns from the beginning of estimation window to the conclusion of event windows as indicated by the aforementioned parameters.

```python
tradedates.to_sql('tradedates', local_sql, index=False, if_exists='replace') 
daily_rets.to_sql('daily_rets', local_sql, index=False, if_exists='replace')  

sqlcode="""    
    select distinct a.*, c.crspday-a.crspday as day, b.date, mktrf, mkt, smb, hml, umd, ret, retrf, exret
    from tradedates as a inner join daily_rets as b
    on a.permno=b.permno and b.date>=estper_beg and b.date<=evtwin_end
    left join crspday as c
    on b.date=c.date
    order by permno, edate, day; 
"""
reg_panel=pd.read_sql_query(sqlcode, local_sql)
```

```python
reg_panel[:3]
```

> We then define an OLS regression function, and calculate parameters over estimation windows for each event

```python
# Define an OLS function with statsmodels module
def regOLS(data, yvar, xvars=None):
    data['const']=1
    y = data[yvar]
    
    if xvars==None:
        X = data['const']
    else: 
        X = pd.concat([data['const'], data[xvars]], axis=1)
    
    result = sm.OLS(y, X).fit() 
    yhat = result.predict(X)
    
    param2 = {'nobs': result.nobs, 'rmse':rmse(y, yhat)} # calc rmse
    param2 = pd.Series(data=param2, index=['nobs', 'rmse'])
    
    res=result.params.append(param2)
    return res
```

```python
#We only keep the daily records over the estimation window prior to S&P inclusion events
est_panel=reg_panel[(reg_panel['estper_beg']<=reg_panel['date'])&(reg_panel['date']<=reg_panel['estper_end'])]
est_panel=est_panel.dropna() #Drop the missing values

#Calcuate total return variance and count Nobs
params_mar=est_panel.groupby(['permno', 'edate']).apply(regOLS, 'exret') #MAR 

#We conduct calculate OLS estimates of various models for each event. 
params_mm=est_panel.groupby(['permno', 'edate']).apply(regOLS, 'ret', 'mkt') #MM
params_ff3=est_panel.groupby(['permno', 'edate']).apply(regOLS, 'retrf', ['mktrf', 'smb', 'hml']) #FF3
params_ff4=est_panel.groupby(['permno', 'edate']).apply(regOLS, 'retrf', ['mktrf', 'smb', 'hml', 'umd']) #FF3+UMD
```

```python
params_mar.head(3)
```

> After calculating the parameters for various models, we utilise the estimated parameters to compute the $\widehat{AR_{i,t}}$ using the above-mentioned formulas with the code below. 

![TimeLine](http://drive.google.com/uc?export=download&id=19d-_l6I4XpNx4hojknc-j0pC4dOURTCV)


```python
car_panel=reg_panel[(reg_panel['day']>=pre) & (reg_panel['day']<=pos)]

car_panel.to_sql('car_panel', local_sql, if_exists='replace') 
params_mar.to_sql('params_mar', local_sql, if_exists='replace') 
params_mm.to_sql('params_mm', local_sql, if_exists='replace') 
params_ff3.to_sql('params_ff3', local_sql, if_exists='replace') 
params_ff4.to_sql('params_ff4', local_sql, if_exists='replace') 

sqlcode="""      
    select distinct a.permno, a.edate, a.day, a.ret, b.nobs as nobs, 

    /*Market Adjusted Model's AR*/
    a.exret as abret0, b.rmse*b.rmse as var0, 

    /*Market Model's AR*/
    a.ret-c.const-c.mkt*a.mkt as abret1, c.rmse*c.rmse as var1, 

    /*Fama French 3 Factor's AR*/
    a.retrf-d.const-d.mktrf*a.mktrf-d.smb*a.smb-d.hml*a.hml as abret2, d.rmse*d.rmse as var2, 

    /*Fama French 3 Factor +USD 's AR*/
    a.retrf-e.const-e.mktrf*a.mktrf-e.smb*a.smb-e.hml*a.hml-e.umd*a.umd as abret3, e.rmse*d.rmse as var3

    from car_panel as a inner join params_mar as b
    on a.permno=b.permno and a.edate=b.edate
    inner join params_mm as c
    on a.permno=c.permno and a.edate=c.edate
    inner join params_ff3 as d
    on a.permno=d.permno and a.edate=d.edate
    inner join params_ff4 as e
    on a.permno=e.permno and a.edate=e.edate

    /*minest (Minimum of non-missing returns required for estimation)*/
    where b.nobs > {} ; 
""".format(minest)

abrets_panel=pd.read_sql_query(sqlcode, local_sql)
```

> Finally, we then calculate the CAR over the event window. To quantify statistical significance, classical student's t statistics are no longer valid, since, by design, each abnormal return may not have the same variance accross different stocks. we employ a method proposed by Patell (1976) that calculates t-values for CARs using standardiszed abnormal return (by idiosyncratic risk). 

> $SCAR(s,t)\approx CAR(s,t)/\sqrt{(t-s+1) \times var(\epsilon_i)}$

> where $\epsilon_i$ is the residual from the risk model estimation for stock i and t-s+1 is the event window length.


```python
# OPTIONAL: You can reset Event Windows Here:
CAR_from=0;                       #Beginning of event window of interest                                 
CAR_to=0;                         #End of event window of interest                                       
evtwin=CAR_to-CAR_from + 1;       #length of event window in trading days 
```

```python
abrets_panel.to_sql('abrets_panel', local_sql, if_exists='replace') 

sqlcode=""" 
    select distinct permno, edate, nobs, 
    sum(abret0) as car0, sum(abret0)/sqrt( {} *var0) as scar0, /*evtwin*/
    sum(abret1) as car1, sum(abret1)/sqrt( {} *var1) as scar1, /*evtwin*/
    sum(abret2) as car2, sum(abret2)/sqrt( {} *var2) as scar2, /*evtwin*/
    sum(abret3) as car3, sum(abret3)/sqrt( {} *var3) as scar3  /*evtwin*/
    from abrets_panel
    where {} <=day /*CAR_fro*/ and day<= {} /*CAR_to*/
    group by permno, edate;
""".format(evtwin,evtwin,evtwin,evtwin,CAR_from,CAR_to)

cars_panel=pd.read_sql_query(sqlcode, local_sql)
```

> The statistic score suggested by Patell(1976) is defined as following:

> $t_{Patell}\approx (t-s+1) \times mean[SCAR(s,t)]$

```python
cars_panel.dropna().to_sql('cars_panel', local_sql, if_exists='replace') 

sqlcode=""" 
    select "MAR" as Model, avg(car0) as car, sum(car0>0) as pos, sum(car0<=0) as neg,
    avg(scar0)*sqrt(count(scar0)) as tpatel
    from cars_panel
    UNION
    select "MM" as Model, avg(car1) as car, sum(car1>0) as pos, sum(car1<=0) as neg,
    avg(scar1)*sqrt(count(scar1)) as tpatel
    from cars_panel
    UNION
    select "FF" as Model, avg(car2) as car, sum(car2>0) as pos, sum(car2<=0) as neg,
    avg(scar2)*sqrt(count(scar2)) as tpatel
    from cars_panel
    UNION
    select "FFM" as Model, avg(car3) as car, sum(car3>0) as pos, sum(car3<=0) as neg,
    avg(scar3)*sqrt(count(scar3)) as tpatel
    from cars_panel;
"""
cars_stats=pd.read_sql_query(sqlcode, local_sql)

#Relabeling the variables
cars_stats = cars_stats.rename(columns={'car': 'CAR('+str(CAR_from)+','+str(CAR_to)+')', 'tpatel': 'Patell t'})
```
