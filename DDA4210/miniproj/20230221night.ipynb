```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from collections import Counter
from sklearn.neural_network import MLPClassifier
from sklearn.ensemble import AdaBoostClassifier
```

```python
#read the data
train = pd.read_csv("train.csv")
test = pd.read_csv("test.csv")
X_train = train[['feature_1', 'feature_2', 'feature_3', 'feature_4']]
y_train = train['label']
X_test_final = test[['feature_1', 'feature_2', 'feature_3', 'feature_4']]
```

```python
#feature_1
train_feat1 = X_train["feature_1"]
test_feat1 = X_test_final["feature_1"]
sns.kdeplot(train_feat1, shade = True, color='r', label = 'train')
sns.kdeplot(test_feat1, shade = True, color='b', label = 'test')
plt.xlabel('Feature')
plt.legend()
plt.title("Feature1")
plt.show()
```

```python
#feature_2
train_feat2 = X_train["feature_2"]
test_feat2 = X_test_final["feature_2"]
sns.kdeplot(train_feat2, shade = True, color='r', label = 'train')
sns.kdeplot(test_feat2, shade = True, color='b', label = 'test')
plt.xlabel('Feature')
plt.legend()
plt.title("Feature2")
plt.show()
```

```python
#feature_3
train_feat3 = X_train["feature_3"]
test_feat3 = X_test_final["feature_3"]
sns.kdeplot(train_feat3, shade = True, color='r', label = 'train')
sns.kdeplot(test_feat3, shade = True, color='b', label = 'test')
plt.xlabel('Feature')
plt.legend()
plt.title("Feature3")
plt.show()
```

```python
#feature_4
train_feat4 = X_train["feature_4"]
test_feat4 = X_test_final["feature_4"]
sns.kdeplot(train_feat4, shade = True, color='r', label = 'train')
sns.kdeplot(test_feat4, shade = True, color='b', label = 'test')
plt.xlabel('Feature')
plt.legend()
plt.title("Feature4_original")
plt.show()
```

```python
from imblearn.over_sampling import SMOTE
```

```python
oversample = SMOTE(random_state=42)
X_train, y_train = oversample.fit_resample(X_train, y_train)
print(Counter(y_train))
```

```python
#Change the distribution of feature 4
miu1 = X_train["feature_4"].mean()
sigma1 = X_train["feature_4"].std()
miu2 = X_test_final["feature_4"].mean()
sigma2 = X_test_final["feature_4"].std()
train_feat4_normal = ((X_train["feature_4"]-miu1)/sigma1)*sigma2+miu2
test_feat4 = X_test_final["feature_4"]
sns.kdeplot(train_feat4_normal, shade = True, color='r', label = 'train')
sns.kdeplot(test_feat4, shade = True, color='b', label = 'test')
plt.xlabel('Feature')
plt.legend()
plt.title("Feature4_normalized")
plt.show()
```

```python
X_train["feature_4"] = train_feat4_normal
```

```python
X_train["feature_5"] = X_train["feature_4"]
for i in range(len(X_train)):
    if X_train["feature_4"][i] <=30:
        X_train["feature_5"][i] = 0
    else:
        X_train["feature_5"][i] = 1
```

```python
X_test_final["feature_5"] = X_test_final["feature_4"]
for i in range(len(X_test_final)):
    if X_test_final["feature_4"][i] <=30:
        X_test_final["feature_5"][i] = 0
    else:
        X_test_final["feature_5"][i] = 1
```

```python
clf_Adaboost = AdaBoostClassifier()
clf_Adaboost.fit(X_train,y_train)
y_pred_final = clf_Adaboost.predict(X_test_final)
```

```python
to_submit = []
for eid, pred in zip(test['example_id'].values, y_pred_final):
    to_submit.append({"example_id": eid, "prediction": pred})
to_submit = pd.DataFrame(to_submit)
to_submit
```
