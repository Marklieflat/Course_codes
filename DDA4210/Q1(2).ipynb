```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data=pd.read_excel('/Users/bhchen/Desktop/2023_MCM-ICM_Problems/Problem_C_Data_Wordle.xlsx',header=1)
data = data.drop(['Unnamed: 0'], axis = 1)
data.sort_values(by='Date',inplace=True)
data.reset_index(inplace=True,drop=True)
data
```

```python
interval = 5
```

```python
data["mean_of_win"] = data['Number of  reported results'].rolling(interval).mean()
groups = []
for i in range(data.shape[0]):
    if i%interval==1:
        groups.append(1)
    else:
        groups.append(0)
```

```python
data['#windows'] = groups
data_new = data[data['#windows']==1]
data_new.dropna(inplace=True)
data_new.sort_values(by='Date')
data_new.reset_index(inplace=True,drop=True)
data_new
```

```python
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LSTM
from keras.models import Sequential, load_model

from sklearn.preprocessing import MinMaxScaler
```

```python
dataWin = data_new.mean_of_win
datasWin = dataWin.astype('float32')
sc = MinMaxScaler(feature_range=(0, 1))
dataWin = sc.fit_transform(dataWin.values.reshape(-1, 1))
dataWin
```

```python
def gen_XY(originalData, step):
    X = []
    Y = []
    for i in range(len(originalData) - step):
        a = originalData[i:(i + step)]
        X.append(a)
        Y.append(originalData[i + step])
    return np.array(X), np.array(Y)
```

```python
x, y = gen_XY(dataWin, 1)
x = np.reshape(x, (x.shape[0], x.shape[1], 1))
len(y)
```

```python
model = Sequential()
model.add(LSTM(4, input_shape=(None,1)))
model.add(Dense(1))
model.compile(loss='mean_squared_error', optimizer='adam')
model.fit(x, y, epochs=100, batch_size=1, verbose=2)
```

```python
y_pred = model.predict(x)
y_pred_inv = sc.inverse_transform(y_pred)
y_inv = sc.inverse_transform(y)
```

```python
from sklearn.metrics import mean_squared_error
```

```python
def score(y, y_hat):
    #MSE
    print("MSE :")
    print(np.sqrt(mean_squared_error(y, y_hat)))
```

```python
score(y, y_pred)
score(y_inv, y_pred_inv)
```

```python
plt.plot(data_new['Date'].values[:-1],y_inv, label='observed data')
plt.plot(data_new['Date'].values[:-1],y_pred_inv, label='LSTM')
plt.xlabel('Date')
plt.ylabel('Number of reported results')
plt.title( 'Number of reported results fitting')
plt.legend()
plt.savefig('./Q1/Number of reported results fitting.jpg') 
plt.show()
```

```python
times = 60/interval
x_latest = y[-1]

list1 = []
list1.append(x_latest)
while len(list1) <= times:
    x_latest = list1[-1].reshape((-1, 1, 1))
    y_pred = model.predict(x_latest, verbose=0)
    list1.append(y_pred)
    
sc.inverse_transform(np.array([i.reshape(-1,1)[:,0].tolist() for i in list1]))
```
