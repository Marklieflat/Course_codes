```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
import math
from sklearn.preprocessing import StandardScaler
```

```python
data=pd.read_csv(r'D:\Code Library\DDA4210\variable_results.csv')
data = data.drop(['Unnamed: 0'], axis = 1)
data
```

```python
data.columns
```

```python
vec_x = data[['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i',
       'j', 'k', 'l', 'm', 'n', 'o', 'p', 'q', 'r', 's', 't', 'u', 'v', 'w',
       'x', 'y', 'z']]
vec_x = vec_x.to_numpy()
```

```python
attribute = ['vowel_cnt','isholiday', 'DT', 'EX', 'FW', 'IN', 'JJ', 'JJR',
       'JJS', 'MD', 'NN', 'NNP', 'NNS', 'PRP$', 'RB', 'RBR', 'VB', 'VBD',
       'VBG', 'VBN', 'VBP', 'VBZ','num of char']
output = ['1 try', '2 tries', '3 tries', '4 tries',
       '5 tries', '6 tries', '7 or more tries (X)']
len(attribute)
```

```python
x0 = data[attribute]
y = data[output]
x0
```

```python
from sklearn.manifold import TSNE
tsne = TSNE(n_components = 3)
x1 = tsne.fit_transform(x0)
x1
```

```python
y
```

```python
scaler = StandardScaler()
scaler.fit(y)
y = scaler.transform(y)
y
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.multioutput import RegressorChain
from sklearn.neural_network import MLPRegressor
from sklearn.metrics import r2_score
from sklearn import ensemble
from sklearn.multioutput import MultiOutputRegressor 
from sklearn.metrics import mean_squared_error
```

```python
def score(y, y_hat):
    result1 = r2_score(y, y_hat, multioutput='raw_values')
    result2 = r2_score(y, y_hat, multioutput='uniform_average')
    return result1, result2
```

```python
train_x, test_x, train_y, test_y = train_test_split(x1 , y, test_size=0.1)
```

```python
train_x, test_x, train_y, test_y = train_test_split(vec_x , y, test_size=0.1, random_state = 2023)
```

```python
model_rf = RegressorChain(RandomForestRegressor(max_depth=7, n_estimators=100, min_samples_split=4))
model_rf.fit(train_x,train_y)
print("Random forest:")
train_y_hat = pd.DataFrame(model_rf.predict(train_x))
test_y_hat = pd.DataFrame(model_rf.predict(test_x))
#print(train_y_hat)
#print(test_y_hat)

print(model_rf.score(train_x,train_y))
print(model_rf.score(test_x,test_y))
```

```python
params = {'n_estimators': 100, 'max_depth': 7, 'min_samples_leaf': 4}

estimator = MultiOutputRegressor(ensemble.RandomForestRegressor(**params))
estimator.fit(train_x,train_y)
mse = mean_squared_error(test_y, estimator.predict(test_x))
print("MSE: %.4f" % mse)
print(estimator.score(test_x,test_y))
```
