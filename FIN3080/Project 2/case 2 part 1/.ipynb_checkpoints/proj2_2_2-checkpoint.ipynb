```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt
```

```python
df_begin= pd.read_excel('个股月回报.xlsx')
```

```python
df_begin =df_begin.rename(columns ={'Stkcd':'Symbol','Trdmnt':'TradingDate'})
df_clear = df_begin.drop((df_begin[df_begin.Symbol.str[0]=='没']).index,axis=0)
df_clear = df_clear.drop((df_clear[df_clear.Symbol.str[0]=='证']).index,axis=0)
df2=df_clear.reset_index(drop=True)
df2
```

```python
df2['Date'] =df2['TradingDate'].astype('datetime64[ns]')
df2
```

```python
df4 = df2.groupby(['Symbol'])
df4.head()
DF =pd.DataFrame()
for i in df4:
    a = i[1].reset_index(drop = True)
    date_range = pd.date_range(start = "20060101 ", end = "20220331", freq='MS')
    a = a.set_index('Date').reindex(index =date_range)
    a['Mretnd-1'] = a['Mretnd'].shift(1)
    a = a.where(a.notnull(),None)    
    c = a.isnull().T.any()
    c = c.loc[c == True]
    columns = c.index.tolist()   
    for column in columns:
        if a.loc[column,'Mretnd-1'] != None:
            a.loc[column,'TradingDate'] = str(column)[:7]
            a.loc[column,'Symbol'] = i[0]
    a = a.drop(['Mretnd'],axis = 1)    
    a = a.dropna()
    DF = pd.concat([DF,a],axis=0,ignore_index=True) 
DF
```

```python
df_ri = pd.read_excel('个股月回报.xlsx')
df_ri.head()
```

```python
df_ri = df_ri.drop([0,1])
```

```python
df_ri = df_ri.reset_index(drop = True)
df_ri =df_ri.rename(columns = {'Trdmnt':'TradingDate','Stkcd':'Symbol'})
df_ri
```

```python
df_ri_Mr = DF.merge(df_ri)
df_ri_Mr
```

```python
df_ri_Mr = df_ri_Mr.dropna(axis = 0,how = 'any')
df_ri_Mr
```

```python
df_ri_Mr = df_ri_Mr[df_ri_Mr['TradingDate'].str[:4]> '2006' ] 
df_ri_Mr
```

```python
def quan(df,col):   
    q1 = df[col].quantile(0.1)
    q2 = df[col].quantile(0.2)
    q3 = df[col].quantile(0.3)
    q4 = df[col].quantile(0.4)
    q5 = df[col].quantile(0.5)
    q6 = df[col].quantile(0.6)
    q7 = df[col].quantile(0.7)
    q8 = df[col].quantile(0.8)
    q9 = df[col].quantile(0.9)
    df.loc[df[col]<=q1, 'Group'] = 1
    df.loc[(df[col]>q1) & (df[col]<=q2), 'Group'] = 2
    df.loc[(df[col]>q2) & (df[col]<=q3), 'Group'] = 3
    df.loc[(df[col]>q3) & (df[col]<=q4), 'Group'] = 4
    df.loc[(df[col]>q4) & (df[col]<=q5), 'Group'] = 5
    df.loc[(df[col]>q5) & (df[col]<=q6), 'Group'] = 6
    df.loc[(df[col]>q6) & (df[col]<=q7), 'Group'] = 7
    df.loc[(df[col]>q7) & (df[col]<=q8), 'Group'] = 8
    df.loc[(df[col]>q8) & (df[col]<=q9), 'Group'] = 9
    df.loc[(df[col]>q9), 'Group'] = 10
    return df
df_g = df_ri_PB.groupby('TradingDate').apply(quan,'Mretnd-1')
df_g.loc[(df_g['Group']==1.0)&(df_g['TradingDate']=='2007-02')]
```

```python
rm = pd.read_excel('沪深300.xlsx')
rm = rm.drop([0,1])
rm = rm.drop(['Indexcd'],axis = 1)
rm = rm.rename(columns ={'Month':'TradingDate'})
rm
```

```python
Erm = rm.mean()
Erm
```

```python
DF_all = pd.merge(df_g1,rm,on=['TradingDate'],how = 'inner')
DF_all
```

```python
rf = pd.read_excel('MBK_SHIBORM.xlsx')
rf = rf.drop([0,1])
rf = rf[rf['Term'] == '1天']
rf = rf.reset_index(drop = True)
rf['TradingDate'] = rf['SgnDate'].str[:7]
del rf['Term']
del rf['SgnDate']
rf['Shibor'] = rf['Shibor']/1200
rf
```

```python
rf =rf.groupby(['TradingDate']).mean()
rf
```

```python
DF_all = pd.merge(DF_all,rf,on=['TradingDate'],how = 'inner')
DF_all =DF_all.reset_index(drop = True)
DF_all
```

```python
df_all = df_all.astype({'Idxrtn':float,'Shibor':float})
```

```python
import statsmodels.api as sm
alpha = pd.Series(dtype = 'float')
for i in range(1,11):
    num = str(i)
    x1 = df_all['Idxrtn']-df_all['Shibor']
    Y = df_all[i]-df_all['Shibor']
    X = sm.add_constant(x1)
    results = sm.OLS(Y,X).fit()
    alpha[num] = results.params['const']

alpha.plot.bar(figsize = (10,6))
plt.axhline(Erm[0],color = 'green',lw = 2,ls ='--')
plt.axhline(0,color = 'black',lw = 2)
plt.grid()
```

```python
plt.figure(figsize=(10,6))
for i in range(1,11):
    df_all['plus']=df_all[float(i)]+1
    df_cumulative=df_all.plus.cumprod()-1
    plt.plot(df_all.TradingDate,df_cumulative)
```
