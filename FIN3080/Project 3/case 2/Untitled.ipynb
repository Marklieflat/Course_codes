```python
df_ab = pd.DataFrame()
for i in df_TRD1:
    a = i[1]
    a = a.merge(rm)
    a['abnormal'] = a['Dretnd']- a['Cdretmdeq']
    a = a.merge(state1, on = ['Stkcd','Trddt'], how ='left')
    num = a[(a['Annodt'].isnull() == False)].index.tolist()
    a = a.drop(['Dretnd','Cdretmdeq','Annodt'],axis = 1)
    ab_all = pd.DataFrame()
    for j in num:
        ab = a.loc[j-120:j+120]
        ab['TrDate'] = a.Trddt[j]
        ab = ab.drop(['Trddt'],axis = 1)
        if len(ab) >= 200:
            ab['index'] = range(-len(ab)//2+1,len(ab)//2+1)
            ab['abnormal'] = ab['abnormal'].cumsum()   #累加后的
            ab_all = ab_all.append(ab)   
    df_ab = df_ab.append(ab_all)
# df_ab = df_ab.reset_index(drop = True)
df_ab
```

```python
EPS = pd.read_excel('FI_T9.xlsx')
EPS
```

```python
EPS1 = EPS.drop([0,1])
EPS1 = EPS1[EPS1['Typrep'] == 'A']
EPS1 = EPS1.drop(['Typrep','F090101C','F090102B','F090102C','F090103B','F090103C','F090104B','F090104C'],axis = 1)
EPS1 = EPS1[(EPS1['Accper'].str[5:] == '06-30') | (EPS1['Accper'].str[5:] == '12-31')]
EPS1 = EPS1.reset_index(drop = True)
EPS1 = EPS1.groupby(['Stkcd'])
EPS1.head()
```

```python
SUE = pd.DataFrame()
for i in EPS1:
    a = i[1]
    a['F090101B-2'] = a['F090101B'].shift(2)
    a['UE'] = a['F090101B'] - a['F090101B-2']
    a['sigma'] = a['UE'].rolling(4).std()
    a['SUE'] = a['UE']/a['sigma']
    a = a.drop(['F090101B-2','F090101B','UE','sigma'], axis = 1)
    a = a.dropna()
    SUE = pd.concat([SUE,a],axis=0,ignore_index=True) 
SUE
```

```python
def quan(df,col):   
    q1 = df[col].quantile(0.1)
    q2 = df[col].quantile(0.2)
    q3 = df[col].quantile(0.3)
    q4 = df[col].quantile(0.4)
    q5 = df[col].quantile(0.5)
    q6 = df[col].quantile(0.6)
    q7 = df[col].quantile(0.7)
    q8 = df[col].quantile(0.8)
    q9 = df[col].quantile(0.9)
    df.loc[df[col]<=q1, 'Group'] = '01'
    df.loc[(df[col]>q1) & (df[col]<=q2), 'Group'] = '02'
    df.loc[(df[col]>q2) & (df[col]<=q3), 'Group'] = '03'
    df.loc[(df[col]>q3) & (df[col]<=q4), 'Group'] = '04'
    df.loc[(df[col]>q4) & (df[col]<=q5), 'Group'] = '05'
    df.loc[(df[col]>q5) & (df[col]<=q6), 'Group'] = '06'
    df.loc[(df[col]>q6) & (df[col]<=q7), 'Group'] = '07'
    df.loc[(df[col]>q7) & (df[col]<=q8), 'Group'] = '08'
    df.loc[(df[col]>q8) & (df[col]<=q9), 'Group'] = '09'
    df.loc[(df[col]>q9), 'Group'] = '10'
    return df
df_q = SUE.groupby('Accper').apply(quan,'SUE')
df_q
```

```python
df_q = df_q.rename(columns = {'Accper':'TrDate'})
df_all = df_ab.merge(df_q, on = ['Stkcd','TrDate'])
df_all = df_all.drop(['SUE'],axis = 1)
df_all
```

```python
df_port = df_all.groupby(['TrDate','Group','index']).mean()
df_port
```

```python
df_port1 = df_port.drop(['Stkcd'],axis = 1)
df_port1
```

```python
df_port2 = df_port1.groupby(['Group','index']).mean()
df_port2 = df_port2.reset_index()
df_port2
```

```python
df_p = pd.pivot_table(df_port2,index = 'index',columns = 'Group',values = 'abnormal')
df_p
```
